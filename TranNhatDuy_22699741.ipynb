```python
#9.	(0.75đ) Vẽ đồ thị Pie để so sánh phân bố các loại thời tiết trong năm.
thoitiet=df.groupby(['Thời_tiết']).size().reset_index(name = 'Số Lượng')
thoitiet
```

```python

plt.pie(thoitiet['Số Lượng'], labels=thoitiet['Thời_tiết'], autopct='%1.1f%%', colors=['blue', 'orange'])
plt.title('Tỉ lệ thời tiết')
plt.show
```

```python
#13.	(0.75đ) Thống kê để hoàn thành bảng số liệu dưới đây.
newData = df.groupby(['Khu_vực']).mean()
khuvuc = newData[['Nhiệt_độ','Lượng_mưa','AQI']].reset_index()
khuvuc
```

```python
# Multibar

# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
NhietDo = khuvuc['Nhiệt_độ']
LuongMua = khuvuc['Lượng_mưa']
AQI = khuvuc['AQI']

# Set position of bar on X axis
br1 = np.arange(len(NhietDo))             
br2 = [x + barWidth for x in br1]       
br3 = [x + barWidth for x in br2]       

# Make the plot
plt.bar(br1, NhietDo, color ='r', width = barWidth,
        edgecolor ='grey', label ='Nhiệt Độ')
plt.bar(br2, LuongMua, color ='g', width = barWidth,
        edgecolor ='grey', label ='Lượng Mưa')
plt.bar(br3, AQI, color ='b', width = barWidth,
        edgecolor ='grey', label ='AQI')

# Adding Xticks
plt.xlabel(['Bắc Bộ','Nam Bộ','Trung Bộ','Tây Nguyên'], fontweight ='bold', fontsize = 15)
plt.ylabel('Stats', fontweight ='bold', fontsize = 15)

plt.legend()
plt.show()
```
