# KHÁM PHÁ DỮ LIỆU

Dữ liệu chứa trong file `tips.csv` chứa thông tin tiền tip của khách hàng tại một nhà hàng. Dữ liệu có các thông tin sau:

- `total_bill`: tổng hoá đơn (USD)
- `tip`: tiền tip/ tiền boa (USD)
- `sex`: giới tính người thanh toán hóa đơn
- `smoker`: có hút thuốc hay không
- `day`: ngày trong tuần
- `time`: thời gian dùng bữa(buổi sáng, buổi trưa, buổi tối)
- `size`: số lượng người tham gia

Bạn hãy thực hiện thống kê mô tả để tìm hiểu về hành vi tip của khách hàng tại nhà hàng

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/tips.csv",header=0)
```

## HIỂU DỮ LIỆU

1. Đọc dữ liệu
2. Xoá cột `Unnamed: 0`
3. Cho biết dữ liệu có bao nhiêu dòng, bao nhiêu thuộc tính.
4. Trong các thuộc tính, thuộc tính nào là thuộc tính phân loại, thuộc tính nào là thuộc tính định lượng
5. Có bao nhiêu nam, bao nhiêu nữ. Tỷ lệ bao nhiêu?
6. Có bao nhiêu người hút thuốc. Tỷ lệ bao nhiêu?
7. Ngày nào trong tuần khách hàng hay đến dùng bữa nhất. Tỷ lệ bao nhiêu?
8. Nhà hàng phục vụ bữa ăn nào nhiều nhất (Breakfast, Lunch, Dinner)?
9. Nhóm khách hàng dùng bữa tại nhà hàng nhiều nhất là bao nhiêu người?
10. Trung bình trong một bữa ăn nhà hàng phục vụ mấy khách?

```python
#1. Đọc dữ liệu
df
df.head()
```

```python
#2. Xoá cột Unnamed: 0
df.drop(columns=["Unnamed: 0"],inplace=True)
df.head()
```

```python
# 3. Cho biết số dòng và số thuộc tính
df.shape
```

```python
#4. Phân loại thuộc tính
phan_loai = ["sex", "smoker", "day", "time"]
dinh_luong = ["total_bill", "tip", "size"]

print("Thuộc tính phân loại:", phan_loai)
print("Thuộc tính định lượng:", dinh_luong)
```

```python
#5. Số lượng nam, nữ và tỷ lệ
dem_gioi_tinh = df["sex"].value_counts()

sl_nam = dem_gioi_tinh["Male"]
sl_nu = dem_gioi_tinh["Female"]
sL_tong = len(df["sex"])

tl_nam = sl_nam / sL_tong * 100
tl_nu = sl_nu / sL_tong * 100

print(f"Số lượng nam là {sl_nam} chiếm {tl_nam}% ")
print(f"Số lượng nữ là {sl_nu} chiếm {tl_nu}% ")
```

```python
#6 Số lượng & tỷ lệ người hút thuốc
dem_nguoi_hut_thuoc = df["smoker"].value_counts()
sl_ng_hut = dem_nguoi_hut_thuoc["Yes"]
sl_ng_ko_hut = dem_nguoi_hut_thuoc["No"]
sL_tong = len(df["smoker"])

tl_nguoi_hut = sl_ng_hut / sL_tong * 100

print(f"Số người hút thuốc: {sl_ng_hut}, chiếm {tl_nguoi_hut}%")
```

```python
#7. Ngày khách đến đông nhất, tỷ lệ 

dem_ngay = df["day"].value_counts()
ngay_dong_nhat = dem_ngay.idxmax()
sL_tong = len(df["day"])

ty_le_ngay_dong_nhat = dem_ngay.max() / sL_tong * 100

print(f"Ngày khách đến đông nhất: {ngay_dong_nhat}, chiếm {ty_le_ngay_dong_nhat}%")
```

```python
#8. Bữa ăn được phục vụ nhiều nhất
dem_bua_an = df["time"].value_counts()
bua_an_nhieu_nhat = dem_bua_an.idxmax()
sL_tong = len(df["time"])

tl_bua_an_nhieu_nhat = dem_bua_an.max() / sL_tong * 100

print(f"Bữa ăn phổ biến nhất: {bua_an_nhieu_nhat}, chiếm {tl_bua_an_nhieu_nhat}%")
```

```python
#9. Nhóm khách hàng đông nhất

nhom_khach_hang = df["size"].value_counts()
nhom_dong_nhat = nhom_khach_hang.idxmax()

print(f"Nhóm khách hàng đông nhất là {nhom_dong_nhat} người.")
```

```python
#10. Trung bình số khách trong một bữa ăn
trung_binh_khach = df["size"].mean()

print(f"Trung bình số khách trong một bữa ăn: {trung_binh_khach} khách.")
```

## YẾU TỐ ẢNH HƯỞNG ĐẾN TIỀN TIP

Trong phần này, chúng ta thử tìm hiểu về tiền tip tại nhà hàng và các yếu tố có thể ảnh hưởng đến nó

1. Tỷ lệ nhận được tiền tip của nhà hàng là bao nhiêu?
2. Tỷ lệ tiền tip trên tổng hóa đơn là bao nhiêu?
3. Giữa nam và nữ ai tip nhiều tiền hơn, ai tip thường xuyên hơn?
4. Giữa người hút thuốc và người không hút thuốc ai tip nhiều tiền hơn, ai tip thường xuyên hơn?
5. Thống kê tổng số tiền tip nhận được theo ngày.

```python
#1. Tỷ lệ nhận được tiền tip của nhà hàng là bao nhiêu?
sl_co_tip = (df["tip"] > 0).sum()
tl_tip = sl_co_tip / len(df["tip"]) * 100

print(f"Tỷ lệ hóa đơn có tip: {tl_tip}%")
```

```python
#2. Tỷ lệ tiền tip trên tổng hóa đơn là bao nhiêu?
tong_tip = df["tip"].sum()
tong_hoa_don = df["total_bill"].sum()
ty_le_tip = (tong_tip / tong_hoa_don) * 100

print(f"Tỷ lệ tiền tip trên tổng hóa đơn: {ty_le_tip}%")
```

```python
#3. Giữa nam và nữ ai tip nhiều tiền hơn, ai tip thường xuyên hơn?
#Tính tổng tiền tip trung bình theo giới tính.
tip_theo_sex = df.groupby("sex")["tip"].mean()
tip_theo_tan_suat = df.groupby("sex")["tip"].apply(lambda x: (x > 0).mean() * 100)

print("Trung bình tiền tip theo giới tính:")
print(tip_theo_sex)

print("\nTần suất tip theo giới tính (% hóa đơn có tip):")
print(tip_theo_tan_suat)
```

=> Vậy trung bình tiền tip của nam nhiều hơn

```python
#4. Giữa người hút thuốc và người không hút thuốc ai tip nhiều tiền hơn, ai tip thường xuyên hơn?
tip_theo_nguoi_hut = df.groupby("smoker")["tip"].mean()
tan_suat_tip_theo_tt = df.groupby("smoker")["tip"].apply(lambda x: (x > 0).mean() * 100)

print("Trung bình tiền tip theo trạng thái hút thuốc:")
print(tip_theo_nguoi_hut)

print("\nTần suất tip theo trạng thái hút thuốc (% hóa đơn có tip):")
print(tip_theo_nguoi_hut)
```

=> Người hút thuốc sẽ tip nhiều hơn và thường xuyên hơn

```python
#5. Thống kê tổng số tiền tip nhận được theo ngày.
tip_theo_ngay = df.groupby("day")["tip"].sum()

print("Tổng số tiền tip theo ngày:")
print(tip_theo_ngay)
```

## Với Matplotlib

Bạn hãy trực quan với thư viện matplotlib

```python
import matplotlib.pyplot as plt
#1. Vẽ đồ thị thể hiện tổng tiền tip theo ngày
tip_by_day = df.groupby("day")["tip"].sum()

plt.figure(figsize=(8, 5))
plt.bar(tip_by_day.index, tip_by_day.values, color="skyblue", edgecolor="black")
plt.title("Tổng tiền tip theo ngày")
plt.xlabel("Ngày trong tuần")
plt.ylabel("Tổng tiền tip")
plt.show()
```

```python
#2. Vẽ đồ thị thể hiện tổng tiền tip theo bữa ăn (time)
tip_by_time = df.groupby("time")["tip"].sum()

plt.figure(figsize=(6, 5))
plt.bar(tip_by_time.index, tip_by_time.values, color="salmon", edgecolor="black")
plt.title("Tổng tiền tip theo bữa ăn")
plt.xlabel("Bữa ăn")
plt.ylabel("Tổng tiền tip")
plt.show()
```

```python
#3. Vẽ đồ thị histogram của tổng hóa đơn (total_bill)
plt.figure(figsize=(8, 5))
plt.hist(df["total_bill"], bins=20, color="lightgreen", edgecolor="black", alpha=0.7)
plt.title("Phân bố tổng hóa đơn")
plt.xlabel("Tổng hóa đơn")
plt.ylabel("Số lượng")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
#4. Vẽ đồ thị boxplot của tổng hóa đơn (total_bill)
plt.figure(figsize=(6, 5))
plt.boxplot(df["total_bill"], vert=True, patch_artist=True, boxprops=dict(facecolor="gold"))
plt.title("Boxplot của tổng hóa đơn")
plt.ylabel("Tổng hóa đơn")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
#5. Vẽ đồ thị để kiểm tra sự liên quan giữa tổng hóa đơn và tiền tip

plt.figure(figsize=(8, 5))
plt.scatter(df["total_bill"], df["tip"], color="purple", alpha=0.5, edgecolors="black")
plt.title("Mối quan hệ giữa tổng hóa đơn và tiền tip")
plt.xlabel("Tổng hóa đơn")
plt.ylabel("Tiền tip")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
#6. So sánh số tiền mà khách hàng đã chi giữa các ngày trong tuần

hd_theo_ngay = [df[df["day"] == d]["total_bill"] for d in df["day"].unique()]

plt.figure(figsize=(8, 5))
plt.boxplot(hd_theo_ngay, labels=df["day"].unique(), patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("So sánh số tiền khách hàng chi giữa các ngày trong tuần")
plt.xlabel("Ngày trong tuần")
plt.ylabel("Tổng hóa đơn")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

## Với Seaborn

Bạn hãy tìm hiểu và trực quan với thư viện seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
#1. Vẽ đồ thị thể hiện tổng tiền tip theo ngày
plt.figure(figsize=(8, 5))
sns.barplot(x="day", y="tip", data=df, estimator=sum, palette="coolwarm")
plt.title("Tổng tiền tip theo ngày")
plt.xlabel("Ngày trong tuần")
plt.ylabel("Tổng tiền tip")
plt.show()
```

```python
#2. Vẽ đồ thị thể hiện tổng tiền tip theo bữa ăn (time)
plt.figure(figsize=(6, 5))
sns.barplot(x="time", y="tip", data=df, estimator=sum, palette="magma")
plt.title("Tổng tiền tip theo bữa ăn")
plt.xlabel("Bữa ăn")
plt.ylabel("Tổng tiền tip")
plt.show()
```

```python
#3. Vẽ đồ thị histogram của tổng hóa đơn (total_bill)

plt.figure(figsize=(8, 5))
sns.histplot(df["total_bill"], bins=20, kde=True, color="blue")
plt.title("Phân bố tổng hóa đơn")
plt.xlabel("Tổng hóa đơn")
plt.ylabel("Số lượng")
plt.show()
```

```python
#4. Vẽ đồ thị boxplot của tổng hóa đơn (total_bill)
plt.figure(figsize=(8, 5))
sns.histplot(df["total_bill"], bins=20, kde=True, color="blue")
plt.title("Phân bố tổng hóa đơn")
plt.xlabel("Tổng hóa đơn")
plt.ylabel("Số lượng")
plt.show()
```

```python
#5. Vẽ đồ thị để kiểm tra sự liên quan giữa tổng hóa đơn và tiền tip
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["total_bill"], y=df["tip"], alpha=0.7)
sns.regplot(x=df["total_bill"], y=df["tip"], scatter=False, color="red")  # Thêm đường hồi quy
plt.title("Mối quan hệ giữa tổng hóa đơn và tiền tip")
plt.xlabel("Tổng hóa đơn")
plt.ylabel("Tiền tip")
plt.show()
```

```python
#6. So sánh số tiền mà khách hàng đã chi giữa các ngày trong tuần
plt.figure(figsize=(8, 5))
sns.boxplot(x="day", y="total_bill", data=df, palette="Set2")
plt.title("So sánh số tiền khách hàng chi giữa các ngày trong tuần")
plt.xlabel("Ngày trong tuần")
plt.ylabel("Tổng hóa đơn")
plt.show()
```
