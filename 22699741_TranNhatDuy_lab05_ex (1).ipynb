**Bài 01:** Một bài thi trắc nghiệm gồm 10 câu hỏi, mỗi câu có 4 phương án trả lời trong đó chỉ có một phương án đúng. Bạn hãy: 

a. Một sinh viên không học bài làm bài bằng cách chọn ngẫu nhiên một phương án cho mỗi câu hỏi. Bạn hãy dùng hàm random để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra.

b. Giả sử mỗi câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm. Tính xác suất để sinh viên này được 4 điểm

c. Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)

```python
import numpy as np
import math as math

# **Bài 01:** Một bài thi trắc nghiệm gồm 10 câu hỏi, mỗi câu có 4 phương án trả lời trong đó chỉ có một phương án đúng. Bạn hãy: 

# a. Một sinh viên không học bài làm bài bằng cách chọn ngẫu nhiên một phương án cho mỗi câu hỏi. Bạn hãy dùng hàm random để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra.

def bai01a():
    n = 10
    p = 1/4
    for i in range(10):
        print(np.random.binomial(n, p))
        # np.random.binomial(n, p) 
        # trả về số lần thành công trong n lần thử, với xác suất thành công là p
print("Câu 1: Số câu đúng của sinh viên trong 10 lần kiểm tra:")
bai01a()

# b. Giả sử mỗi câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm. Tính xác suất để sinh viên này được 4 điểm

def bai01b():
    n = 10
    p = 1/4
    x = 4 # câu dúng
    y = 2 # câu sai
    q = 1 - p
    # xác suất để sinh viên được 4 điểm
    P = math.comb(n, x) * (p**x) * (q**(n-x))
    # comb(n, k) trả về số cách chọn k phần tử từ n phần tử
    print(P)

print("Câu 2: Xác suất để sinh viên được 4 điểm:")
bai01b()
# c. Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)

def bai01c():
    n = 10
    p = 1/4
    q = 1 - p
    E = n * p
    Var = n * p * q
    print(E, Var)

print("Câu 3: Kỳ vọng và phương sai của số câu trả lời đúng:")
bai01c()
```

**Bài 02:** Một trung tâm bưu điện nhận được 3 cuộc gọi điện thoại mỗi phút. Tính xác suất để trung tâm nhận được 1 cuộc gọi, 2 cuộc gọi, 3 cuộc gọi trong vòng một phút biết số cuộc gọi trong một phút có phân phối Poisson.

```python
# **Bài 02:** Một trung tâm bưu điện nhận được 3 cuộc gọi điện thoại mỗi phút. Tính xác suất để trung tâm nhận được 1 cuộc gọi, 2 cuộc gọi, 3 cuộc gọi trong vòng một phút biết số cuộc gọi trong một phút có phân phối Poisson.

def bai02():
    lambda_ = 3
    k = 1
    k2 = 2
    k3 = 3
    # xác suất để trung tâm nhận được 1 cuộc gọi
    P1 = (lambda_**k) * np.exp(-lambda_) / math.factorial(k)
    # xác suất để trung tâm nhận được 2 cuộc gọi
    P2 = (lambda_**k2) * np.exp(-lambda_) / math.factorial(k2)
    # xác suất để trung tâm nhận được 3 cuộc gọi
    P3 = (lambda_**k3) * np.exp(-lambda_) / math.factorial(k3)
    print("Xác suất để trung tâm nhận được 1 cuộc gọi: ", P1)
    print("Xác suất để trung tâm nhận được 2 cuộc gọi: ", P2)
    print("Xác suất để trung tâm nhận được 3 cuộc gọi: ", P3)

bai02()
```

```python
# **Bài 03:** Trọng lượng (đơn vị gam) của một loại trái cây có phân phối chuẩn với µ = 500 (gam) gam và $σ^2 = 16 (gam^2)$. Trái cây thu hoạch được phân loại theo trọng lượng như sau:

# Loại 1: trên 505 gam

# Loại 2: từ 495 – 505 gam

# Loại 3: dười 495 gam

# a. Hãy tính tỷ lệ của mỗi loại.
from scipy.stats import norm
import matplotlib.pyplot as plt
def bai03a():
    mu = 500
    sigma = 4

    # Xác suất loại 1: P(X > 505) = 1 - P(X ≤ 505)
    P1 = 1 - norm.cdf(505, mu, sigma)

    # Xác suất loại 2: P(495 ≤ X ≤ 505) = P(X ≤ 505) - P(X ≤ 495)
    P2 = norm.cdf(505, mu, sigma) - norm.cdf(495, mu, sigma)

    # Xác suất loại 3: P(X < 495) = P(X ≤ 495)
    P3 = norm.cdf(495, mu, sigma)

    print("Tỷ lệ của mỗi loại:")
    print(f"Loại 1 (> 505g): {P1:.4f} ({P1*100:.2f}%)")
    print(f"Loại 2 (495g - 505g): {P2:.4f} ({P2*100:.2f}%)")
    print(f"Loại 3 (< 495g): {P3:.4f} ({P3*100:.2f}%)")

bai03a()

# b. Bạn hãy mô phỏng lấy mẫu  20 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu

def bai03b():
    np.random.seed(42)  # Để kết quả có thể lặp lại
    sample_20 = np.random.normal(500, 4, 20)  # Lấy mẫu 20 trái cây từ phân phối chuẩn

    # Vẽ histogram
    plt.hist(sample_20, bins=7, color='skyblue', edgecolor='black', alpha=0.7)
    plt.axvline(np.mean(sample_20), color='red', linestyle='dashed', linewidth=2, label=f'Mean: {np.mean(sample_20):.2f}')
    plt.axvline(500, color='green', linestyle='dashed', linewidth=2, label="True Mean: 500")
    
    plt.title("Histogram của 20 trái cây lấy mẫu")
    plt.xlabel("Trọng lượng (gam)")
    plt.ylabel("Tần suất")
    plt.legend()
    plt.show()

    print(f"Trung bình mẫu: {np.mean(sample_20):.2f}")
    print(f"Độ lệch chuẩn mẫu: {np.std(sample_20, ddof=1):.2f}")  # ddof=1 để tính s (ước lượng từ mẫu)

bai03b()

# c. Bạn hãy mô phỏng lấy 100 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu

def bai03c():
    np.random.seed(42)
    sample_100 = np.random.normal(500, 4, 100)  # Lấy mẫu 100 trái cây

    # Vẽ histogram
    plt.hist(sample_100, bins=10, color='orange', edgecolor='black', alpha=0.7)
    plt.axvline(np.mean(sample_100), color='red', linestyle='dashed', linewidth=2, label=f'Mean: {np.mean(sample_100):.2f}')
    plt.axvline(500, color='green', linestyle='dashed', linewidth=2, label="True Mean: 500")
    
    plt.title("Histogram của 100 trái cây lấy mẫu")
    plt.xlabel("Trọng lượng (gam)")
    plt.ylabel("Tần suất")
    plt.legend()
    plt.show()

    print(f"Trung bình mẫu: {np.mean(sample_100):.2f}")
    print(f"Độ lệch chuẩn mẫu: {np.std(sample_100, ddof=1):.2f}")  # ddof=1 để tính s (ước lượng từ mẫu)

bai03c()

# d. Từ hai cách lấy mẫu ở câu b, c bạn có nhận xét gì không?

# Khi tăng kích thước mẫu từ 20 lên 100, giá trị trung bình và độ lệch chuẩn của mẫu cũng tăng lên. Điều này cho thấy mẫu lớn hơn sẽ có giá trị trung bình và độ lệch chuẩn gần với giá trị trung bình và độ lệch chuẩn của quần thể hơn.
```

**Bài 04:** Mô phỏng tung một con xúc sắc cân đối đồng chất 5000 lần. Dựa vào giá trị mô phỏng, bạn hãy tìm các giá trị xác suất dưới đây:

a. Xác suất để  số chấm xuất hiện là 4

b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4

c. Giả sử biết số chấm xuất hiện lớn hơn hoặc bằng 4. Hãy tìm xác suất để mặt 6 chấm xuất hiện.

```python
def bai04():
    n = 5000  # Số lần tung xúc xắc
    rolls = np.random.choice([1, 2, 3, 4, 5, 6], size=n, replace=True)  # Mô phỏng tung xúc xắc

    # a. Xác suất để số chấm xuất hiện là 4
    P_4 = np.sum(rolls == 4) / n

    # b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4
    P_ge_4 = np.sum(rolls >= 4) / n

    # c. Xác suất để mặt 6 xuất hiện, biết rằng số chấm ≥ 4 (định lý Bayes)
    P_6_given_ge_4 = np.sum((rolls == 6) & (rolls >= 4)) / np.sum(rolls >= 4)

    print(f"Xác suất số chấm là 4: {P_4:.4f}")
    print(f"Xác suất số chấm ≥ 4: {P_ge_4:.4f}")
    print(f"Xác suất số 6 xuất hiện biết rằng số chấm ≥ 4: {P_6_given_ge_4:.4f}")

bai04()
```

b. Vẽ đồ thị boxplot

```python

def bai05a():
    np.random.seed(42)  
    data = np.random.binomial(n=50, p=0.7, size=5000) 

    # Tính giá trị thống kê
    min_val = np.min(data)
    max_val = np.max(data)
    Q1 = np.percentile(data, 25) 
    Q2 = np.median(data)          
    Q3 = np.percentile(data, 75)  

    print(f"Min: {min_val}")
    print(f"Max: {max_val}")
    print(f"Q1 (25%): {Q1}")
    print(f"Q2 (Median - 50%): {Q2}")
    print(f"Q3 (75%): {Q3}")

bai05a()

def bai05b():
    np.random.seed(42)
    data = np.random.binomial(n=50, p=0.7, size=5000)

    plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor="skyblue"))
    plt.title("Boxplot của phân phối nhị thức (n=50, p=0.7)")
    plt.xlabel("Giá trị")
    plt.show()

bai05b()


def bai05c():
    np.random.seed(42)
    # np.ramdom.binomial: mô phỏng phân phối nhị thức
    # n: số lần thử, p: xác suất thành công
    data = np.random.binomial(n=50, p=0.7, size=5000)

   
    # unique: giá trị duy nhất, counts: số lần xuất hiện
    unique, counts = np.unique(data, return_counts=True)

    
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.bar(unique, counts, color="orange", alpha=0.7)
    plt.xlabel("Giá trị")
    plt.ylabel("Tần số")
    plt.title("Đồ thị tần số của phân phối nhị thức")

    
    plt.subplot(1, 2, 2)
    plt.bar(unique, counts / len(data), color="blue", alpha=0.7)
    plt.xlabel("Giá trị")
    plt.ylabel("Tần suất")
    plt.title("Đồ thị tần suất của phân phối nhị thức")

    plt.tight_layout()
    plt.show()

bai05c()

```

**Bài 6**

Giả sử X là biến ngẫu nhiên rời rạc

a. Hãy cho biết ý nghĩa câu lệnh sau: `binom.pmf(4, 10, 0.6)`

b. Giả sử X có phân phối nhị thức `X ~ B(10, 0.6)`. Hãy tính các giá trị sau:  
$P(X \leq 5), P(X \le 5), P(X \ge 4), P(X = 5) $

c. Hãy cho biết ý nghĩa câu lệnh sau: `poisson.pmf(4, 3)`

d. Giả sử X có phân phối Poisson `X ~ P(3)`. Hãy tính các giá trị sau:  
$P(X \leq 5), P(X \le 5), P(X \ge 4), P(X = 5) $

```python
#a. Câu lệnh binom.pmf(4, 10, 0.6) trong SciPy được sử dụng để tính xác suất khối (PMF - Probability Mass Function) của biến ngẫu nhiên có phân phối nhị thức. Trong trường hợp này, xác suất để có k thành công trong n lần thử với xác suất thành công là p. Trong trường hợp này, xác suất để có 4 thành công trong 10 lần thử với xác suất thành công là 0.6 là 0.2508226560000002.

# b. Giả sử X có phân phối nhị thức `X ~ B(10, 0.6)`. Hãy tính các giá trị sau:  
# $P(X \leq 5), P(X \le 5), P(X \ge 4), P(X = 5) $
from scipy.stats import binom
def cau6b():
    n = 10  # số lần thử
    p = 0.6  # xác suất thành công

    # P(X <= 5)
    P_X_leq_5 = binom.cdf(5, n, p)

    # P(X >= 4) = 1 - P(X < 4) = 1 - P(X <= 3)
    P_X_geq_4 = 1 - binom.cdf(3, n, p)

    # P(X = 5)
    P_X_eq_5 = binom.pmf(5, n, p)

    print(f"P(X <= 5): {P_X_leq_5:.4f}")
    print(f"P(X >= 4): {P_X_geq_4:.4f}")
    print(f"P(X = 5): {P_X_eq_5:.4f}")
print("Câu 6b:")
cau6b()

# c. Hãy cho biết ý nghĩa câu lệnh sau: `poisson.pmf(4, 3)`

# Câu lệnh poisson.pmf(4, 3) trong SciPy được sử dụng để tính xác suất khối (PMF - Probability Mass Function) của biến ngẫu nhiên có phân phối Poisson. Trong trường hợp này, xác suất để có 4 sự kiện xảy ra trong một khoảng thời gian với trung bình 3 sự kiện xảy ra là 0.16803135574154085.

# d. Giả sử X có phân phối Poisson `X ~ P(3)`. Hãy tính các giá trị sau:  
# $P(X \leq 5), P(X \le 5), P(X \ge 4), P(X = 5) $
from scipy.stats import poisson
def cau6d():
    lambda_ = 3  # trung bình
    k = 5

    # P(X <= 5)
    P_X_leq_5 = poisson.cdf(k, lambda_)

    # P(X >= 4) = 1 - P(X < 4) = 1 - P(X <= 3)
    P_X_geq_4 = 1 - poisson.cdf(3, lambda_)

    # P(X = 5)
    P_X_eq_5 = poisson.pmf(5, lambda_)

    print(f"P(X <= 5): {P_X_leq_5:.4f}")
    print(f"P(X >= 4): {P_X_geq_4:.4f}")
    print(f"P(X = 5): {P_X_eq_5:.4f}")
print("Câu 6d:")
cau6d()
```

```python
# **Bài 8**

# Một công ty sản xuất làm ra các chi tiết máy với độ dài các chi tiết có phân phối chuẩn với kỳ vọng là 3000 mm và độ lệch chuẩn là 3 mm. Yêu cầu cho các chi tiết máy này là phải có chiều dài từ 2993 mm đến 3007 mm. Tính tỷ lệ sản phẩm lỗi của công ty

def bai08():
    mu = 3000
    sigma = 3

    # P(2993 <= X <= 3007)
    P = norm.cdf(3007, mu, sigma) - norm.cdf(2993, mu, sigma)

    print(f"Tỷ lệ sản phẩm lỗi: {1 - P:.4f}")

bai08()
```

Giả sử điểm thi TOEIC là một biến ngẫu nhiên có phân phối chuẩn với kỳ vọng là 500, độ lệch chuẩn là 8. Bạn hãy:

a. Mô phỏng bằng cách phát sinh ngẫu nhiên điểm thi TOIEC của một nhóm gồm 50 người.

b. Tính điểm trung bình và độ lệch chuẩn về điểm của nhóm.

c. Tìm miền giá trị, và miền phân vị (IQR) của nhóm

d. Cho biết tỷ lệ đạt trên 450 điểm của nhóm. 

e. Vẽ đồ thị histogram của nhóm và so sánh với đồ thị hàm mật độ của phân phối chuẩn với kỳ vọng là 500 và độ lệch chuẩn là 8.

f. Thay vì mô phỏng 50 người, hãy mô phỏng nhóm với 100 người thi. Bạn có nhận xét gì?

```python
# **Bài 9**

# Giả sử điểm thi TOEIC là một biến ngẫu nhiên có phân phối chuẩn với kỳ vọng là 500, độ lệch chuẩn là 8. Bạn hãy:

# a. Mô phỏng bằng cách phát sinh ngẫu nhiên điểm thi TOIEC của một nhóm gồm 50 người.

def bai09a():
    np.random.seed(42)
    data = np.random.normal(500, 8, 50)
    return data

print("Câu 9a:")
print(bai09a())

# b. Tính điểm trung bình và độ lệch chuẩn về điểm của nhóm.

def bai09b():
    data = bai09a()
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    return mean, std

print("Câu 9b:")
print("Mean", "Std")
print(bai09b())


# c. Tìm miền giá trị, và miền phân vị (IQR) của nhóm

def bai09c():
    data = bai09a()
    min_val = np.min(data)
    max_val = np.max(data)
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    return min_val, max_val, IQR

print("Câu 9c:")
print("Min Value", "Max Value", "IQR")
print(bai09c())

# d. Cho biết tỷ lệ đạt trên 450 điểm của nhóm. 

def bai09d():
    data = bai09a()
    #P(X > 450)
    P = 1 - norm.cdf(450, np.mean(data), np.std(data, ddof=1))
    return P

print("Câu 9d:")
print(bai09d())

# e. Vẽ đồ thị histogram của nhóm và so sánh với đồ thị hàm mật độ của phân phối chuẩn với kỳ vọng là 500 và độ lệch chuẩn là 8.

def bai09e():
    data = bai09a()
    x = np.linspace(450, 550, 1000)
    y = norm.pdf(x, 500, 8)

    plt.hist(data, bins=10, color='skyblue', edgecolor='black', alpha=0.7, density=True)
    plt.plot(x, y, color='red', label='PDF')
    plt.title("Histogram và PDF của điểm thi TOEIC")
    plt.xlabel("Điểm")
    plt.ylabel("Tần suất")
    plt.legend()
    plt.show()

print("Câu 9e:")
bai09e()

# f. Thay vì mô phỏng 50 người, hãy mô phỏng nhóm với 100 người thi. Bạn có nhận xét gì?

def bai09f():
    np.random.seed(42)
    data = np.random.normal(500, 8, 100)
    x = np.linspace(450, 550, 1000)
    y = norm.pdf(x, 500, 8)

    plt.hist(data, bins=10, color='orange', edgecolor='black', alpha=0.7, density=True)
    plt.plot(x, y, color='red', label='PDF')
    plt.title("Histogram và PDF của điểm thi TOEIC")
    plt.xlabel("Điểm")
    plt.ylabel("Tần suất")
    plt.legend()
    plt.show()

print("Câu 9f:")
bai09f()

# Khi tăng kích thước mẫu từ 50 lên 100, phân phối của dữ liệu sẽ trở nên trơn hơn và gần với phân phối chuẩn hơn. Điều này cho thấy mẫu lớn hơn sẽ có phân phối gần với phân phối chuẩn hơn.
```

**Bài 10**

Viết một chương trình để mô phỏng việc tung đồng xu và đánh giá phương sai của giá trị mô phỏng của biến ngẫu nhiên X (tung được mặt `Head`) đối với các số lần tung khác nhau. Giả sử  đồng xu phải cân bằng. 

Đối với mỗi N từ 10; 40; 90; 160; 250; 490; 640; 810; 1000, hãy ước tính giá trị của X bằng cách mô phỏng số lần tung đó. 

Bạn nên chạy mỗi mô phỏng 100 lần và sử dụng tập hợp các ước tính để đánh giá phương sai của ước tính x của bạn. Vẽ biểu đồ phương sai này so với N, bạn có nhận xét gì không?

```python
# **Bài 10**

# Viết một chương trình để mô phỏng việc tung đồng xu và đánh giá phương sai của giá trị mô phỏng của biến ngẫu nhiên X (tung được mặt `Head`) đối với các số lần tung khác nhau. Giả sử  đồng xu phải cân bằng. 

# Đối với mỗi N từ 10; 40; 90; 160; 250; 490; 640; 810; 1000, hãy ước tính giá trị của X bằng cách mô phỏng số lần tung đó. 

# Bạn nên chạy mỗi mô phỏng 100 lần và sử dụng tập hợp các ước tính để đánh giá phương sai của ước tính x của bạn. Vẽ biểu đồ phương sai này so với N, bạn có nhận xét gì không?

def bai10():
    np.random.seed(42)
    N = [10, 40, 90, 160, 250, 490, 640, 810, 1000]
    variances = []

    for n in N:
        means = []
        for _ in range(100):
            data = np.random.choice([0, 1], size=n, replace=True)
            means.append(np.mean(data))
        variances.append(np.var(means))

    plt.plot(N, variances, marker='o', color='orange')
    plt.title("Phương sai của ước lượng X theo N")
    plt.xlabel("N")
    plt.ylabel("Phương sai")
    plt.show()

bai10()

# Phương sai của ước lượng X giảm khi kích thước mẫu tăng. Điều này cho thấy ước lượng X sẽ trở nên chính xác hơn khi có nhiều dữ liệu hơn.
```
