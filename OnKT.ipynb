```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc file dữ liệu
df = pd.read_csv('/mnt/data/diem_thi_thpt_2022.csv')

# Hiển thị thông tin tổng quan về dữ liệu
print("Thông tin dữ liệu:")
print(df.info())

# Xử lý giá trị thiếu
df.fillna(df.mean(numeric_only=True), inplace=True)
df.dropna(subset=['tentinh'], inplace=True)

# Tạo cột tổng điểm
df['TongDiem'] = df[['toan', 'Ngữ Văn', 'Ngoại Ngữ']].sum(axis=1)

# Xác định học sinh có điểm tổng cao nhất và thấp nhất
max_score = df.loc[df['TongDiem'].idxmax()]
min_score = df.loc[df['TongDiem'].idxmin()]
print("\nHọc sinh có tổng điểm cao nhất:")
print(max_score)
print("\nHọc sinh có tổng điểm thấp nhất:")
print(min_score)

# Phân loại học sinh
def classify(score):
    if score >= 27:
        return 'Xuất sắc'
    elif score >= 21:
        return 'Khá'
    else:
        return 'Trung bình'

df['XepLoai'] = df['TongDiem'].apply(classify)

# Thống kê môn Toán và Ngữ Văn
stats = df[['toan', 'Ngữ Văn']].describe()
print("\nThống kê môn Toán và Ngữ Văn:")
print(stats)

# Hệ số tương quan giữa Toán và Ngoại Ngữ
correlation = df[['toan', 'Ngoại Ngữ']].corr().iloc[0, 1]
print(f"\nHệ số tương quan giữa Toán và Ngoại Ngữ: {correlation:.2f}")

# Phân tích điểm trung bình theo tỉnh
province_avg = df.groupby('tentinh')[['toan', 'Ngữ Văn', 'Ngoại Ngữ']].mean()
highest_province = province_avg.mean(axis=1).idxmax()
lowest_province = province_avg.mean(axis=1).idxmin()
print(f"\nTỉnh có điểm trung bình cao nhất: {highest_province}")
print(f"Tỉnh có điểm trung bình thấp nhất: {lowest_province}")

# Vẽ biểu đồ histogram môn Toán
plt.figure(figsize=(8, 5))
sns.histplot(df['toan'], bins=20, kde=True, color='blue')
plt.xlabel('Điểm Toán')
plt.ylabel('Số lượng học sinh')
plt.title('Phân bố điểm Toán')
plt.show()

# Vẽ boxplot so sánh điểm Toán theo xếp loại
plt.figure(figsize=(8, 5))
sns.boxplot(x='XepLoai', y='toan', data=df, palette='Set2')
plt.xlabel('Xếp loại')
plt.ylabel('Điểm Toán')
plt.title('So sánh điểm Toán theo xếp loại')
plt.show()

# Vẽ biểu đồ Bar thể hiện số lượng học sinh theo nhóm xếp loại
plt.figure(figsize=(6, 4))
df['XepLoai'].value_counts().plot(kind='bar', color=['green', 'orange', 'red'])
plt.xlabel('Xếp loại')
plt.ylabel('Số lượng học sinh')
plt.title('Số lượng học sinh theo nhóm xếp loại')
plt.xticks(rotation=0)
plt.show()
```
