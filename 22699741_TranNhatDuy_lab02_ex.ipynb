## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

```python
import numpy as np
from numpy import random

arr = random.randint(1000, size=100)
print("Mảng: ",arr)
print("10 phần tử đầu tiên",arr[:10])
print("5 phần tử cuối cùng",arr[-5:])
print("Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100: ",arr[(arr%5==0) & (arr>100)])
```

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

```python
arr[arr<100].size
```

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

```python
print("Tính tổng mảng: ", arr.sum())
print("Tìm giá trị trung bình: ",arr.mean())
print("Tìm giá trị trung vị: ",np.median(arr))
print("Tìm Q1, Q2, Q3: ", np.quantile(arr,[0.25,0.5,0.75]))
print("Tìm phân vị 20 (percentile 20): ", np.quantile(arr,0.2))
```

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

```python
from numpy import random 

print(random.choice(arr,size=(5,10),replace=False)) 

```

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

```python
values, counts = np.unique(arr, return_counts=True)

most_frequency_value = np.max(counts)
indices = np.where(counts == most_frequency_value)[0]

for i in range(1, len(indices)):
  print(values[i], end=" ")
  if (i % 10 == 0):
    print()
    
```

### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

```python
# This cell contains output. Don't remove!
```

```python
# YOUR CODE HERE
new_array = np.linspace(0, np.pi,1000)

y1 = np.sin(new_array)
y2 = np.cos(new_array)

plt.plot(new_array,y1)
plt.plot(new_array,y2)
plt.show()
```

### Câu 5: Vẽ các đồ thị con

```python
plt.subplot(2,1,1)
plt.plot(new_array, y1)

plt.subplot(2,1,2)
plt.plot(new_array, y2, linestyle='--')

plt.show()
```

### Câu 8: Vẽ histogram

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ histogram các số ngẫu nhiên trên.

```python
# YOUR CODE HERE
x = np.random.normal(size=10000)     

plt.hist(x, bins=100)
plt.xlabel('Giá trị')
plt.ylabel('Tần số')
plt.title('Histogram')

plt.show()
```
